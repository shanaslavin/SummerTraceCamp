```python
# horoscope 
# asks for user's birthday
# gives back zodiac sign and horoscope

import random

def pick_horoscope(fortunes):
    size = len(fortunes)
    return random.randint(0, size-1)
    

def zodiac_sign(month, day):
    if(month == 1):
        if(day <= 19):
            return "Capricorn"
        else:
            return "Aquarius"
    elif(month == 2):
        if(day <= 18):
            return "Aquarius"
        else:
            return "Pisces"
    elif(month == 3):
        if(day <= 20):
            return "Pisces"
        else:
            return "Aries"
    elif(month == 4):
        if(day <= 19):
            return "Aries"
        else:
            return "Taurus"
    elif(month == 5):
        if(day < 20):
            return "Taurus"
        else:
            return "Gemini"
    elif(month == 6):
        if(day <= 20):
            return "Gemini"
        else:
            return "Cancer"
    elif(month == 7):
        if(day <= 22):
            return "Cancer"
        else:
            return "Leo"
    elif(month == 8):
        if(day <= 22):
            return "Leo"
        else:
            return "Virgo"
    elif(month == 9):
        if(day <= 22):
            return "Virgo"
        else:
            return "Libra"
    elif(month == 10):
        if(day <= 22):
            return "Libra"
        else:
            return "Scorpio"
    elif(month == 11):
        if(day <= 21):
            return "Scorpio"
        else:
            return "Sagittarius"
    elif(month == 12):
        if(day <= 21):
            return "Sagittarius"
        else:
            return "Capicorn"
    else:
        return "broken"
    
    
def add_horoscope(fortunes):
    print(fortunes)
    fortune = input("Please enter a new horoscope: ")
    fortunes.append(fortune)
    print(fortunes)

def delete_horoscope(fortunes):
    for index, fortune in enumerate(fortunes):
        print(index, fortune)
    rm = int(input("Which horoscope do you want to delete? "))
    print(rm)
    removed = fortunes.pop(rm)
    print(fortunes)
    
def varify_month(month):
    if(month < 1 or month > 12):
        print("The month must be between 1 and 12")
        month = int(input("What month were you born? Please input a number: "))
        varify_month(month)
    return month

def varify_day(month, day):
    if(month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12):
        if(day > 31 or day < 0):
            print("The day must be smaller than 31")
            day = int(input("What day were you born? Please input a number: "))
            varify_day(month, day)
    elif(month == 2):
        if(day > 29 or day < 0):
            print("The day must be smaller than 28 unless you are born on a leap year")
            day = int(input("What day were you born? Please input a number: "))
            varify_day(month, day)
    else:
        if(day > 30 or day < 0):
            print("The day must be less than 30")
            day = int(input("What day were you born? Please input a number: "))
            varify_day(month, day)
    return day
```

```python
fortunes = ["You will have a good day", "You will die today", "Today is just not your day"]

month = int(input("What month were you born? Please input a number: "))
month = varify_month(month)

day = int(input("What day were you born? Please input a number: "))
day = varify_day(month, day)


zodiac = zodiac_sign(month, day)

in_use = True

while(in_use):
    scope = pick_horoscope(fortunes)
    print("1: Add a new horoscope")
    print("2: Delete a horoscope")
    print("3: Get your horoscope")
    print("4: Exit")
    user = int(input("Please enter a number: "))
    if(user == 1):
        add_horoscope(fortunes)
    elif(user == 2):
        delete_horoscope(fortunes)
    elif(user == 3):
        print("Your zodiac sign is: ", zodiac)
        print("Your horoscope is: ", fortunes[scope])
    else:
        print("Goodbye")
        in_use = False
```
